```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'age':[2,4,6],'height':[20,40,60],'wts':[32,34,38],'gender':['M','F','M']})
df
```

```python
x=df[['age','height','wts']]
#calculating the eign value and eign vector 

evalue,evector=np.linalg.eig(x)
print(evalue)
```

```python
print(evector)
```

```python
pc=np.dot(x,evector)
pc
```

```python
#calculating the pcs
pcs=pd.DataFrame(pc)
pcs
```

```python
total_var=pcs.var()
total_var
```

```python
total_var=total_var.sum()
total_var
```

```python
#calculating the information given by each pc
var_of_each_pc=pd.DataFrame(pcs.var())
var_of_each_pc
#percentage information of each pcs
var_of_each_pc['percentage']=(round((var_of_each_pc[0]/total_var),2)*100)
var_of_each_pc
```

```python
#ploting 
from matplotlib import pyplot as plt
x_axis=[1,2,3]
y_axis=var_of_each_pc['percentage']
plt.plot(x_axis,y_axis)
plt.show()
```

```python
from pydataset import data
```

```python
iris=data('iris')
iris
```

```python
iris['Species'].unique()
```

```python
X=iris[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']]
Y=iris[['Species']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
len(X_train),len(X_test),len(Y_train),len(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#creating a standarzed object
sc1=StandardScaler()
X_train1=sc1.fit_transform(X_train)
X_test1=sc1.transform(X_test)
```

```python
#craete the model
from sklearn.ensemble import RandomForestClassifier

classifer=RandomForestClassifier()
#fitting the object on training data
model1=classifer.fit(X_train1,Y_train)
model1
```

```python
#predicitng on the test data
Y_test['pred_species']=model1.predict(X_test1)
```
