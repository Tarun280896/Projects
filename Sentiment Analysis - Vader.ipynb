```python
#creating a fucntion to pass it on to the data frame
text_polar = lambda text : senti.polarity_scores(text)['compound']
```

Creating a function to create a column based on the polarity score 

```python
def polarity (score):
    if score['Polarity_score'] > 0.8 and score['Polarity_score'] <= 1:
        return 'High Positive'
    elif score['Polarity_score'] > 0.25 and score['Polarity_score'] <= 0.8:
        return 'Positive'
    elif score['Polarity_score'] > -0.25 and score['Polarity_score'] <= 0.25:
        return 'Nutral'
    elif score['Polarity_score'] > -0.8 and score['Polarity_score'] <= -0.25:
        return 'Negitive'
    elif score['Polarity_score'] < -0.25 and score['Polarity_score'] >= -1 :
        return "High Negitive"
    else:
        return "unknown"
```

```python
Senti_DF = pd.DataFrame(columns=['Review','Polarity_type'])
for index,rows in Amazon_data.iterrows():
    Senti_row={
        'Review' : rows['reviews.text'],
        'Polarity_type':polarity(rows)
    }
    Senti_DF=Senti_DF.append(Senti_row,ignore_index=True)
    
```
